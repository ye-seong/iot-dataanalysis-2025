````python
# 기본데이터
arr = np.random.random(50) * 100
arr
````

````python
# 단일차트(라인차트)
plt.plot(arr)
````

````python
# 출력메시지를 삭제하려면
plt.plot(arr)
plt.show()
````

````python
# 히스토리맵
plt.hist(np.random.randn(100))
plt.show()
````

````python
# 산점도
plt.scatter(np.arange(30), np.random.randn(30))
plt.show()
````

````python
# 다중차트 1 - 여러차트를 한꺼번에 출력
fig = plt.Figure()  # 객체 생성

ax1 = fig.add_subplot(2, 2, 1)  # 2x2 차트 중 1번째
ax2 = fig.add_subplot(2, 2, 2)  # 2x2 차트 중 2번째
ax4 = fig.add_subplot(2, 2, 4)  # 2x2 차트 중 4번째

ax1.hist(np.random.randn(100))
ax2.scatter(np.arange(30), np.random.randn(30))
ax4.plot(arr)
plt.show()
````

````python
# 다중차트 2 - 여러차트를 한꺼번에 출력
fig, ax = plt.subplots(2,2,figsize=(15,8))

# color 색상변경, marker 포인트모양 (o, ^, *)
# linestyle -> -, --, :, -.
ax[0][0].hist(np.random.randn(100), color='green')
ax[0][1].scatter(np.arange(30), np.random.randn(30), color='blue', marker='*')
ax[1][0].plot(arr, color='crimson', linestyle='-', marker='o')
ax[1][1].plot(arr, color='orange', linestyle='-.')
plt.show()
````
